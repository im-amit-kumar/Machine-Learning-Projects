```python
import cv2
face_class = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
cam = cv2.VideoCapture(0)
count = 0
while True:
    ret ,img = cam.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_class.detectMultiScale(gray, 1.3, 5)
    
    for x,y,w,h in faces:
        count = count + 1
        #crop_face = gray[y:y+h, x:x+w]
        color_face = img[y:y+h , x:x+w]
        color_face_resize = cv2.resize(color_face,(200,200))
        crop_face_gray = cv2.cvtColor(color_face_resize, cv2.COLOR_BGR2GRAY)
        cv2.imwrite("DataSets/user{}.jpg".format(count),crop_face_gray)
        cv2.imshow("My face",crop_face_gray)
        
    if count > 150:
        break
        
        
    #cv2.imshow("My Face",img)
    #cv2.imshow("My Face 1",gray)
 
   

    if cv2.waitKey(10) ==13:
        break
        
cam.release()        
cv2.destroyAllWindows()    
```

```python
#Train Model with your Faces
```

```python
import cv2
import numpy as np
from os import listdir
path = "DataSets/"

all_images = listdir(path)

training_data = []
labels = np.arange(1,len(all_images)+1)
#sam = 1
for i in all_images:
    img_path = path + i
    if i[-4:] == ".jpg" :
        img = cv2.imread(img_path , cv2.IMREAD_GRAYSCALE)
        Narray = np.asarray(img, dtype = np.uint8)
        training_data.append(Narray)
        #labels.append(sam)
        #sam +=1
        
labels = np.asarray(labels,dtype=np.int32)        
```

```python
labels
```

```python
face_brain = cv2.face_LBPHFaceRecognizer.create()
face_brain.train(training_data, labels)
print("My Face Brain is Trained")
```

```python

import cv2
face_class = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
cam = cv2.VideoCapture(0)
while True:
    ret ,img = cam.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_class.detectMultiScale(gray, 1.3, 5)
#x,y,w,h =faces[0][0], faces[0][1], faces [0][2] , faces[0][3]
    for x,y,w,h in faces:
        face = img[y:y+h , x:x+w]
        color_face = cv2.resize(face,(200 , 200))
        gray_face = cv2.cvtColor(color_face, cv2.COLOR_BGR2GRAY)
        out = face_brain.predict(gray_face)
        print(out)
        if out[1] < 40:
            cv2.rectangle(img , (x,y) , (x+w , y+h),(0,255,0),2)
            cv2.putText(img , "Hi ! Amit",(x,y-15),cv2.FONT_HERSHEY_SIMPLEX,1.2,(0,255,0),3)
        else:
            cv2.rectangle(img , (x,y) , (x+w , y+h),(0,0,255),2)
            cv2.putText(img , "Unknown Sample",(x,y-15),cv2.FONT_HERSHEY_SIMPLEX,1.2,(0,0,255),3)

   
    cv2.imshow("My Face",img)
    #cv2.imshow("My Face 1",gray)
 
   

    if cv2.waitKey(10) ==13:
        break
        
cam.release()        
cv2.destroyAllWindows()    
```
