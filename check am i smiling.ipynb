```python
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.image import img_to_array
import cv2
import numpy as np
from tensorflow.keras.utils import to_categorical
```

```python
positive_images = os.listdir("SMILEs/positives/")
negative_images = os.listdir("SMILEs/negatives/")
```

```python
data =[]
labels =[]
```

```python
for imagePath in positive_images:
    path = "./SMILEs/positives/" + imagePath
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = cv2.resize(image, (28,28))
    image = img_to_array(image)
    data.append(image)
    labels.append("Smilling")
    
    
for imagePath in negative_images:
    try:
        path = "./SMILEs/negatives/" + imagePath
        image = cv2.imread(path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, (28,28))
        image = img_to_array(image)
        data.append(image)
        labels.append("not_Smilling")
    except:
        pass
```

```python
data = np.array(data, dtype = "float") / 255.0
labels = np.array(labels)
```

```python
Le = LabelEncoder().fit_transform(labels)
labels = to_categorical(Le)
```

```python
x_train , x_test , y_train , y_test = train_test_split(data , labels , test_size = 0.2, random_state =12)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Flatten ,Conv2D , MaxPooling2D , Dropout
```

```python
model = Sequential()

model.add(Conv2D(20 , (5,5) , activation = "relu" , input_shape = (28,28,1)))
model.add(MaxPooling2D(pool_size = (2,2) , strides = (2,2)))

model.add(Conv2D(20 , (5,5) , activation = "relu" ))
model.add(MaxPooling2D(pool_size = (2,2) , strides = (2,2)))

model.add(Flatten())
model.add(Dense(500, activation ="relu"))

model.add(Dense(2, activation = "softmax"))

model.compile(loss = "binary_crossentropy" , optimizer = "adam" ,metrics = ["accuracy"])
```

```python

face_class= cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
cam = cv2.VideoCapture(0)
while True:
    ret, img = cam.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_class.detectMultiScale(gray, 1.3 , 5)  

    for x,y,w,h in faces:
        roi = gray[y:y+h, x:x+w]
        roi = cv2.resize(roi, (28,28))
        roi = roi.astype("float")/255
        roi = img_to_array(roi)
        roi = np.expand_dims(roi, axis = 0)
        
        Not, smiling = model.predict(roi)[0]
        Status = ""
        color = ""
        if Not > smiling:
            Status = "Not Smiling"
            color = (0,0,255)
        else:
            Status = "Smiling"
            color = (0,255,0)
            
        print(Status)
        cv2.rectangle(img, (x,y), (x+w, y+h), color, 2)
        cv2.putText(img, Status, (x, y-10), cv2.FONT_HERSHEY_COMPLEX, 0.50, color, 2)

    cv2.imshow("My Face", img)
    if cv2.waitKey(10) == 13:    
        break

cv2.destroyAllWindows()
```
